###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
df.fillna(0,inplace=True)
```

```python
df.head()
```

```python
perfomanc=df['python']+df['mysql']+df['CodeKata Score']+df['python_en']+df['computational_thinking']
df["perfomance"]=perfomanc.round()
a=df.sort_values('perfomance',ascending=False)
a
```

```python
g=a.groupby('Department').max()
g
```

```python
df.to_csv("new.csv")
```

```python
perf=[]

for x in g['perfomance']:
  perf.append(x)
perf
```

```python
d=[]
for x in df['perfomance']:
  d.append(x)
```

```python
name_num=[]
for x in perf:
  for y in df['perfomance']:
    if x==y:
      name_num.append(d.index(x))
name_num
```

```python
new_names=[]
for x in name_num:
  new_names.append(df['Name'][x])

new_names
```

```python
g['Name'].replace(new_names, inplace=True)
g.drop(['python',	'python_en',	'mysql',	'computational_thinking',	'Previous Geekions',	'CodeKata Score',	'Rising'],axis=1 ,inplace=True)
```

```python
g.reset_index(inplace=True)
g['Name']=new_names
g
```

```python
name_dpt=[]
for x in range(len(g)):
  c=g['Name'][x]+": DEPT of :"+g['Department'][x]
  name_dpt.append(c)
name_dpt

g.plot(kind='pie', y='perfomance',figsize=(15,15) ,colors=colors,explode=explode, autopct='%1.1f%%',labels=name_dpt,fontsize=18)
plt.title('Department wise toppers',fontsize=28)
plt.legend(labels=name_dpt, loc='center',prop={'size': 15})
explode = (0.1, 0.1, 0.1)

plt.savefig("/content/Task4/Department wise toppers.png",bbox_inches='tight')
```
